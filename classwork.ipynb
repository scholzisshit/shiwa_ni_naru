````````````````````````python
import numpy as np
import matplotlib.pyplot as plot

fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 80, 70])
color = np.array(["red", "yellow", "orange", "green"])

plot.bar(fruit, quantity, color=color)
plot.xlabel("Fruit")
plot.ylabel("Quantity of Fruit")
plot.title("Fruit vs Fruit quantity")
plot.show()


````````````````````````
