##**Graphs**
It is a visual representation of data that helps us:
* Understand patterns or trends
* Compares the values

###**Common Types of Graph**
* Line Plot
* Bar Graph
* Histogram
* Scatter plot
* Pie Chart
* Box Plot
* Heatmap

**If you want to create a graph, we need to import a library matplotlib**

```python
import matplotlib.pyplot as plt
```

###**Line Plot - Trends over time**
eg. If you want to show how your monthly savings changed over the year

```python
month = ['jan','feb','march','apr','may']
savings = [5000,7000,6000,6500,8000]
# .plot() -> line plot
plt.plot(month,savings,marker = 'o',color = 'red')
plt.title('Monthly Saving')
plt.xlabel('Monthly')
plt.ylabel('Saving in Ruppess')
plt.grid()
plt.show() # Display the plot
```

###**2. Bar Chart (Categorical Comparision)**
eg. Sales in different products

```python
product = ['A','B','C','D'] # Categories
sales = [120,340,210,180]

#.bar() -> Bar graph
plt.bar(product,sales, color = 'green')
plt.show()
```

##**Histogram**
It is used to check the distribution of the data
* eg. How many students scored between 0-11, 10,20

```python
marks = [10,20,22,30,35,40,50,55,60,70,80,90,96,100]
plt.hist(marks,bins = 10,edgecolor = 'black')
plt.show()
```

##**ScatterPlot**
* If we want to check the relation between any two continous data


```python
height = [150,160,165,170,175,180]
weight = [50,55,60,70,75,80]
# .scatter()
plt.scatter(height,weight)
plt.show()
```

###**Pie Chart**
It is also used for comparision of categories (when the Categories are more than 5 then Bar graph is suggested)

```python
brand = ['A','B','C','D']
price = [30,40,50,60]
plt.pie(price,labels=brand,autopct = '%.2f%%',explode=[0,0,0.05,0])
plt.show()
```

###**Horizontal Bar Chart**

```python
brand = ['A','B','C','D']
price = [120,340,210,180]
# .barh()
plt.barh(brand,price,color = 'green')
plt.title('Brand Price')
plt.xlabel('Price')
plt.ylabel('Brand')
plt.show()
```

##**Multi Line Graph**

```python
days = ['mon','tue','wed','thu','fri']
city1 = [30,32,33,35,36] # Temp
city2 = [25,24,27,22,26]

plt.plot(days,city1,label = 'Delhi')
plt.plot(days,city2,label='Banglore')
plt.show()
```

###**Box Plot**
It is used for outlier analysis (Outliers is some extreme values that we have in our data)

```python
Age = [20,40,50,60,140,-40]
plt.boxplot(Age)
plt.show()
```

###**Heatmap**
This graph is used for implementing Correlation and visualize it

**Note**: In Matplotlib we don't have heat map. To implement heatmap we need to use Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.get_dataset_names()
```

Let's implement some of the graph from seaborn

```python
sns.lineplot(data=df,x='size',y='total_bill')
plt.show()
```

###**Bar Plot**

```python
sns.barplot(data=df,x = 'day',y='tip',errorbar=('ci', False), palette='magma'
)
plt.show()
```

###**Countplot**
It counts the frequency

```python
sns.countplot(data = df,x = 'day',palette = 'pastel')
plt.show()
```

###**Histogram**

```python
sns.histplot(data=df,x = 'total_bill',bins = 20,kde=True)
plt.show()
```

###**Box Plot**

```python
sns.boxplot(data = df, y = 'tip')
plt.show()
```

###**Violin Plot**
Box plot + kde

```python
sns.violinplot(data=df,y= 'tip')
plt.show()
```

###**Scatter plot**

```python
sns.scatterplot(data=df,x='tip',y='total_bill')
plt.show()
```

###**Pair Plot**
Multiple scatter plot

```python
sns.pairplot(df)
plt.show()
```
