##**Machine Learning**
It is a part of computer science where the machine learns from the data and give the solution for the new data

**ML Concepts**
* Data -> The informtion that we are passing to my machine
* ALgorithm -> It is a method or step that are used to build models
* Models -> An program that is trained on given algorithm to get answer
* Training -> It is the process on which my model look at the data and learn from it (Both Question and Answer)
* Testing -> After Training, We check if the model gives me the correct answer or not
* Evaluation -> Once, i am done with my predictions, i will check my model prediction if it is close to my actual answer or not
* We will select the best model for the process

##**Types of Machine Learning**
1. Supervised ML: If you have already given answers with the data (Labelled Data)
2. Unsupervised ML: We have unlabelled data (No Target Column (no answer)) on which we will find the patterns and group
3. Reinforcement Learning: Insead of working on dataset, The learning focus on agent behaviour based on action performed in state and observed by the environment

##**Supervised Learning**
It is also being classified into two parts
* Regression - If we have a continuous target column (Like, House price prediction, Stock prediction, Tempreature)
    * Linear Regression
    * Decision Tree Regressor
    * Random Forest Regressor
    * SVR
    * XGBoost Regressor
* Classification - If we have a categorical target column (Yes|No , Dog|Cat)
    * Logistic Regression
    * Decision Tree
    * Random Forest
    * SVC
    * boosting

##**Unsupervised learning**
It learns from unlabelled data where we don't have target (we find patterns and associaitions)
1. Clustering Algorithms - Grouping, similar kind of data points together (K means, Agglomerative, GMM)
2. Dimensionality Reduction - PCA (Principal component analysis)
3. Association Rule Mining Algo - Used in market basket analysis (Apriori Algorithm)
