Merge , concat, join

```python
data = {
    'Name' : ['Aryan','Sakshi','Sonal','Usha'],
    'Marks': [99,95,70,100]
}

df = pd.DataFrame(data)
df
```

```python
df2 = {
    'NewName': ['Vani','Jishnu','Muskan','Rahul'],
    'Marks' : [99,95,70,100],
    'Age': [25,26,27,28]
}

df2 = pd.DataFrame(df2)
df2
```

```python
pd.merge(df,df2,on='Marks') # SQL join
```

```python
pd.concat([df,df2]) # Vertically
```

```python
pd.concat([df,df2],axis=1)
```
